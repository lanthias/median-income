```python
import altair as alt
import pandas as pd
from vega_datasets import data

us_counties = alt.topo_feature(data.us_10m.url, 'states')
columns = [str(year) for year in range(1984, 2018)]

slider = alt.binding_range(min=1984, max=2017, step=1)
select_year = alt.selection_single(name="year", fields=['year'],
                                   bind=slider, init={'year': 2006})
single = alt.selection_multi(fields=['id'])

geo_chart = alt.Chart(us_counties).mark_geoshape(
    stroke='black',
    strokeWidth=0.05,
    cursor='pointer'
).project(
    type='albersUsa'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(fdf, 'fips', columns)
).transform_fold(
    columns, as_=['year', 'value']
).transform_calculate(
    year='parseInt(datum.year)',
).encode(
    tooltip='value:Q',
    color='value:Q',
    opacity=alt.condition(single, alt.value(1), alt.value(0.3))
).add_selection(
    select_year
).transform_filter(
    select_year
).properties(width=700, height=500).add_selection(single)
geo_transform['id'] = geo_transform.fips
chart = alt.Chart(geo_transform).transform_filter(single).encode(x='year:T', y='value', color='State').mark_area(opacity=0.5, line=True, width=10).properties(width=700, height=500)
main = (chart & geo_chart)
main
```
